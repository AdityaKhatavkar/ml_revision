## PCA

It is unsupervised ML problem. It is complex and old techniue.

    Aim of PCA : To reduce the curse of dimensionality.
                 PCA converts high dimensnional data into the low dimensions.

    Benefits of PCA : 
                Faster execution of the algorithm.
                Better visualization.

### Previous Knowledge should know about:

1. Metrices :
 
        Metrices are basically linear transformations, which are applied on any co-ordinate axes results in the transformation of all the points and vectords present in the co-ordinate axes.

    2. Covariance Matrix :
        Covariance matrix is the matrix of the covariance between the variables in a data set.
        Diagonal elements in the cov matrix shows variance of that perticula feature with itself. And rest of the matrix is symmetirc.

3. Eigenvectors :

        Eigen vectors are special vectors whose direction not changes after applying the transformaitons. Their magnitude may change.
        In 2D data there are 2 eign vectors. Similiarly in 3D data there are 3 eign vectors. and so on.

    4. Eigenvalues :
    
        Eigenvalues represent the factors by which an eigenvector is stretched or compressed during the transformation represented by A.

   5. Eigendecomposition of the covariance matrix.
   
        Largest eigenvector of the covariance matrix always points into the direction of largest variance of the data and magnitude of this vector equals to the corresponding eigen value.

## Steps in the PCA
- Make the data mean centric (optional step)
- Find covariance matrix
- Find the eigen vectors and eigen values
- show results
