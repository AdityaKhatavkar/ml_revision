```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
from sklearn.linear_model import Ridge
```

```python
rr0 = Ridge(alpha = 10)
rr0.fit(x,y)
print(rr0.coef_)
print(rr0.intercept_)
```

```python
rr1 = Ridge(alpha = 100)
rr1.fit(x,y)
print(rr1.coef_)
print(rr1.intercept_)
```
