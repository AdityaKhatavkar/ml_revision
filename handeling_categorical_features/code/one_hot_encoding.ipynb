## What is ohe

If data is nominal,
data will not follow intrinsic order

We create columns for each caategory, depending on the occurance of that category we will put 1 or 0 as value for that perticualr tuple

## Dummy Variable Trap

Columns form due to the ohe technique are called as 'Dummy Variables' 

Dummy variable can cause the dataset to have multicollinearity among the features.

Example:

If you observed, the sum of all the columns in ohe is 1. Hence, the dummy variable trap is present in the dataset.

To avoid this dummy variable trap:
1. ohe will create n features for n categories
2. ohe will drop any one (generally first) feature from all dummy features created.
3. it will help to reduce the multicollinearity among the features. 


## OHE using most frequent variables

Feature may contain large number of categories, it will lead to high number of dummy features.

We keep those categories which occures maximum. And create another column say 'other' for rest of the categories.

## Ohe with sklearn

Ohe with sklearn is little tidious task. Becoz,

We can not apply the ohe directly on the dataframe

We have to apply the ohe on required features only, and then have to merge the ohe encoded columns with the rest of the columns.

To fix this we have column transformer in sklearn, we will see it later

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=0)
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first', sparse_output=False, dtype=np.int32)
```

```python
x_train_new = ohe.fit_transform(x_train[['fuel','owner']])
```

```python
x_test_new = ohe.fit_transform(x_test[['fuel','owner']])
```

```python
x_train_new.shape
```

```python
# lets analyze ohe's hyperparameters
'''
ohe = OneHotEncoder(drop='first',sparse=False,dtype=np.int32)

1. drop : 
           To drop or not to drop first column from feature.
           options : 'first' , 'if_binary' , boolean
           default : first

2. sparse_output :
           To return sparse matrix or not.
           if we not apply 'sparse = False', then we have to convert it to array explictly
           x_train_new = ohe.fit_transform(x_train[['fuel'.owner]]).toarray() 

3. dtype : 
            Default numbers are float.
            If we want another dtype, we can specify it.
'''
```

## OHE with top categories

To reduce no. of dummy features, categories with less than threshold occurances we well name them as 'other'

```python
df['brand'].value_counts()
```

```python
threshold = 100
```

```python
replace_brand = df['brand'].value_counts() < threshold
```

```python
replace_brand
```

```python
pd.get_dummies(df['brand'].replace(replace_brand[replace_brand == True].index, 'other'),dtype = int)
```
