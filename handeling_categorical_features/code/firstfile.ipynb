### Ordinal data : 

It is a categorical data where the order is important.

for example: 
- reveies -> good / bad / poor
- quality -> good / bad / poor
- marks -> 90 , 80 , 70

## Nominal data :

It is a categorical data where the order is not important.

for example:

- color -> red , green , blue
- gender -> male , female

### Different techniques are there to handle categorical data.

1. One hot encoding
2. Label encoding
3. Target encoding
4. Ordinal encoding
